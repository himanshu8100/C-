```
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test =train_test_split(X, Y, test_size=0.20, shuffle=True)
```

```
print(X_train.shape)
print(X_test.shape)
```

```
#sigmoid function
def sigmoid_fun(x):
  return (1/(1+np.exp(-x)))

class Logistic_Regression:
  
  #initialization
  def __init__(self, learning_rate= 0.001, iteration =500):
    self.lrn_rate= learning_rate
    self.n_iter= iteration
    self.weight=None
    self.bias=None
 
  def fit(self, X, Y):
    n_datapoints, n_features= X.shape

    #initialize wieghts
    self.weight= np.random.rand(n_features)*0.01
    self.bias= 0

    for _ in range(self.n_iter):
      #forward prop
      z= np.dot(self.weight,X.T)
      y_predicted= sigmoid_fun(z)

      #gradients
      dz= y_predicted- z
      dw= (1/n_datapoints)*(np.dot(X.T,y_predicted-Y))
      db= 1/n_datapoints*(np.sum(y_predicted-Y))
      
      #update values using gradient
      self.weight -= self.lrn_rate * dw
      self.bias -= self.lrn_rate * db

  #prediction  
  def predict(self, X):
    z = np.dot(X, self.weight) + self.bias
    y_predicted= sigmoid_fun(z)
    y_class= np.array([1 if i> 0.5 else 0 for i in y_predicted])
    return y_class
```

```
LR= Logistic_Regression()
```

```
LR.fit(X_train, Y_train)
```

```
Y_out=LR.predict(X_test)
```

```
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_test, Y_out)
```
